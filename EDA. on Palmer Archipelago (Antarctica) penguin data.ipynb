```python
Let me define each term for you in the context of the Palmer Archipelago penguin data:

- Sample Number: It's a unique identifier given to each sample collected from the penguins.
    
- Species: It refers to the specific type of penguin species found in the Palmer Archipelago.
    
- Region: It indicates the geographical area within the Palmer Archipelago where the penguins were observed.
    
- Island: It represents the specific island within the Palmer Archipelago where the penguins were recorded.
    
- Stage: It refers to the developmental stage of the penguins, such as juvenile, sub-adult, or adult.
    
- Individual ID: It's a unique identifier assigned to each individual penguin for tracking purposes.
    
- Clutch Completion: It indicates whether the penguin's clutch (nest) was successfully completed or not.
    
- Date Egg: It represents the date when the egg was laid by the penguin.
    
- Culmen Length (mm): It refers to the length of the upper beak of the penguin, measured in millimeters.
    
- Culmen Depth (mm): It represents the depth of the upper beak of the penguin, measured in millimeters.
    
- Flipper Length (mm): It indicates the length of the penguin's flipper, measured in millimeters.
    
- Body Mass (g): It refers to the weight of the penguin's body, measured in grams.
    
- Delta 15 N (o/oo): It's a measurement used to determine the nitrogen isotope composition in the penguin's diet.
    
- Delta 13 C (o/oo): It's a measurement used to determine the carbon isotope composition in the penguin's diet.

These data points help researchers study various aspects of penguin populations in the Palmer Archipelago. 🐧📊
```

# Lets Detect if any 'OUTLIERS'/'Missing values'

```python
Q3] Detect OUTLIERS/Missing values
```

```python
Q3(a)] What is outlier
ANSWER :- An outlier is a data point that stands out or doesn't fit with rest of the data
```

```python
Q3(b)] what is Missing values
ANSWER :- Missing values refer to the absence of data in certain observations or variables
```

```python
#Detect OUTLIERS/Missing values
sns.boxplot(data = df[['Culmen Length (mm)', 'Body Mass (g)']])
plt.title('Outlier in Culmen Length and Body Mass')
plt.show()
```

```python
# Find the distribution of data

def boxplots(col):
    sns.boxplot(df[col])
    plt.show()
    
for i in list(df.select_dtypes(exclude=["object"]).columns)[1:]:
    boxplots(i)
```
